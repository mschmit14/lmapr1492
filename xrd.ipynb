```python
import numpy as np
from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
mp_key = "bw9PJDXW6ljZz8DyLmXT49ZnMengQUwm"
mp_id = "mp-30460"
```

```python
#récupération de la structure
with MPRester(mp_key) as m:
    struct = m.get_structure_by_material_id(mp_id) 
```

```python
#Computes the XRD pattern of a crystal structure
xrd = XRDCalculator('CuKa',1.54060)

#Calculates the diffraction pattern for a structure
pattern = xrd.get_pattern(struct)


#premier pic de diffraction
print("Premier pic de diffraction:")
print("Indices hkl associés: ", pattern.hkls[0])
print("Intensité = ",pattern.y[0])  
print("2\u03B8(°) = ", pattern.x[0],"°")
print("distance interplan = ",pattern.d_hkls[0])
print("\n")

#deuxième pic de diffraction
print("Deuxième pic de diffraction:")
print("Indices hkl associés: ", pattern.hkls[1])
print("Intensité = ",pattern.y[1])
print("2\u03B8(°) = ", pattern.x[1],"°")
print("distance interplan = ",pattern.d_hkls[1])
print("\n")

#troisième pic de diffraction
print("Troisème pic de diffraction:")
print("Indices hkl associés: ", pattern.hkls[2])
print("Intensité = ",pattern.y[2])   
print("2\u03B8(°) = ", pattern.x[2],"°")
print("distance interplan = ",pattern.d_hkls[2])
print("\n")

xrd.get_plot(struct)
```
